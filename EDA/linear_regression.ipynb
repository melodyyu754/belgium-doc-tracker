```python
import numpy as np


def linear_regression(X, y):
    # Don't forget your docstring!
    """performs the linear perceptron algorithm which takes in an original X matrix, y vector, initial weights vector, learning rate, and maximum iterations and outputs a weight vector that determines the direction and orientation of the decision boundary

    Args:
        X (2d-array): represents a matrix of a bias column with numeric values representative of the x features 
        y (1d-array): represents a vector of labels (-1 or 1)
        w (1d-array): represents an initial weight vector 
        alpha (int): represents the learning rate
        max_iter (int): the maximum number of iterations the regression model should keep going through
    Returns:
        w (1d-array): the final weight vector that determines the direction and orientation of the decision boundary 
    """
    return np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y) #gradient = slope
```
